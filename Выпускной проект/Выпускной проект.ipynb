# Описание проекта: промышленность

Чтобы оптимизировать производственные расходы, металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали. Для этого комбинату нужно контролировать температуру сплава. Ваша задача — построить модель, которая будет её предсказывать.

Заказчик хочет использовать разработанную модель для имитации технологического процесса. Ниже расскажем о деталях этого процесса. Их важно знать, прежде чем генерировать новые признаки.
Описание этапа обработки:
- Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 
- Из сплава выводится сера (этот процесс — десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (от англ. tribe — «масса»).
- Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.
- Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (от англ. slab — «плита»).

В данных наблюдается одно единственное отрицательное значение. Реактивная мощность может принимать отрицательные значения, однако, в данном случае такое значение всего одно и скорее всего является выбросом. Нужно проверить отдельно всю партию.

```python
df_arc_new[df_arc_new['key'] == 2116]
```

В партии это значение единственное. Чтобы не нарушать целостность данных, можно проверить корреляцию между активной энергией и реактивной, и в случае совпадения, рассчитать реактивную мощность по активной.

```python
filtered_data = df_arc_new[(df_arc_new['key'] == 2116) & (df_arc_new['reactive_power'] > 0)]

plt.scatter(filtered_data['active_power'], filtered_data['reactive_power'])

x = sm.add_constant(filtered_data['active_power'])
model = sm.OLS(filtered_data['reactive_power'], x).fit()
plt.plot(filtered_data['active_power'], model.predict(x), color='r')

plt.show()
print(model.summary())
```

Получили необходимые коэффициенты **-0.3398** и **1.3116**. Можем построить зависимость и заменить аномальное значение.

```python
df_arc_new.at[9780, 'reactive_power'] = df_arc_new.iloc[9780].active_power * 1.3116 - 0.3398
```

Добавим несколько новых столбцов для удобства.

Столбец `total_power` будет содержать в себе значение полной мощности, рассчитанной по формуле 

$$
S=\sqrt{P^2 + Q^2}
$$

Столбец `duration_arc` будет содержать в себе полное время нагрева сплава дугой

Столбец `energy` будет содержать в себе полную энергию, затраченную на нагрев

```python
df_arc_new['total_power'] = np.sqrt(np.power(df_arc_new['active_power'], 2) + np.power(df_arc_new['reactive_power'], 2))
df_arc_new['duration_arc'] = (df_arc_new['end_arc_heating'] - df_arc_new['start_arc_heating']).astype('timedelta64[s]')
df_arc_new['energy'] = df_arc_new['total_power'] * df_arc_new['duration_arc']
```

## Нейронная сеть

Отмасштабируем данные для нейронной сети

```python
scaler = StandardScaler()
```

```python
features_train_nn = scaler.fit_transform(features_train)
features_test_nn = scaler.transform(features_test)
```

```python
def r2_metric(y_true, y_pred):
    total_error = tf.reduce_sum(tf.square(tf.subtract(y_true, tf.reduce_mean(y_true))))
    unexplained_error = tf.reduce_sum(tf.square(tf.subtract(y_true, y_pred)))
    r2 = tf.subtract(1.0, tf.divide(unexplained_error, total_error))

    return r2
```

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Dense(64, input_shape=(features_test_nn.shape[1],)))
model.add(tf.keras.layers.Dense(32))
model.add(tf.keras.layers.Dense(1))

model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.Adam(), metrics=['mae', r2_metric])
model.summary()
```

```python
history = model.fit(features_train_nn, target_train, epochs=11, batch_size=32, validation_data=(features_test_nn, target_test))
```

```python
model_results['neural_network'] = [min(history.history['val_mae']), abs(max(history.history['val_r2_metric']))]
```

Самый лучший показатель `r2` наблюдается у модели `Ridge`, что означает, что данная модель хорошо объясняет вариацию зависимой переменной.

Проведем оценку качества лучшей модели `LGBMRegressor` на тестовых данных.

```python
best_predictions = best_lgbm.predict(features_test)
print('MAE LGBMRegressor на тестовых данных:', round(mean_absolute_error(target_test, best_predictions), 2))

if mean_absolute_error(target_test, best_predictions) < mean_absolute_error(target_test, dummy_predictions):
    print('MAE LGBMRegressor меньше, чем на константой модели. Итоговая модель адекватна.')
else:
    print('MAE LGBMRegressor больше, чем на константой модели. Итоговая модель неадекватна.')
```
